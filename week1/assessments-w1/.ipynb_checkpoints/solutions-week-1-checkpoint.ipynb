```python
def payroll(pay_per_hour, hours):
    if((pay_per_hour<0)|(hours<0)):
        raise ValueError('pay_per_hour and hours have to be non-negative numbers')
    return pay_per_hour*hours
```

```python
payroll(7.5,10)
```

```python
def fibonacci(n):
# Sequence where each number in sequnce is the sum of the previous two
### BEGIN SOLUTION
    myarray=[1,1] # First two elements of the sequence
    # iterate from position 2 all the way to n-1
    for i in range(2,n): 
        # create element "i" as sum of previous two elements
        myarray.append(myarray[i-1]+myarray[i-2])
    return myarray
    ### END SOLUTION
```

```python
fibonacci(10)
```

```python
def myfactorial(num):
    '''Calculate the Factorial of the input num
    the output is an int
    
    For instance, "four factorial" is written as "4!" and means 1×2×3×4 = 24. 
    In general, n! ("enn factorial") means the product of all the whole numbers from 1 to n; 
    that is, n! = 1×2×3×...×n.
    '''
    if(type(num)!=int):
        raise VauleError('input is not an integer')
    if(num<=0):
        raise VauleError('input is not an integer')
    '''start your code here, num is a positive integer here'''
    ### BEGIN SOLUTION  
    myfactorial = 1
    for i in range(1,num+1):
        myfactorial = myfactorial * i
    return myfactorial
    ### END SOLUTION
    '''end your code'''
```

```python
def find_ID(email):
    """Check if duke address and if so, return user name from the input email address.  Otherwise return None   
    """
    ### BEGIN SOLUTION
    user = email.split("@") # create list by splitting by "@"
    if(user[1]=="duke.edu"):
        return user[0] # return first element
    else:
        return None # return none
    ### END SOLUTION
```

```python
def cubics(n):
    """Compute the cubics of numbers from 1 to n, such that the 
    ith element of the returned list equals i^3.
    
    """
    ### BEGIN SOLUTION
    if n < 1:
        raise ValueError("Invalid Input")
    cubed = list(range(n))
    cubed[0]=1
    for i in range(2,n+1):
        cubed[i-1]=((i)**3)
    return cubed
    ### END SOLUTION
```

```python
cubics(10)
```

```python
def sum_of_cubics(n):
    """Compute the sum of the cubics of numbers from 1 to n."""
    ### BEGIN SOLUTION
    return sum(cubics(n))
    ### END SOLUTION
```

```python
sum_of_cubics(10)
```

```python
def Collatz(n):
    """For input n (must be a positive integer), compute a list of numbers in reverse order, that form a Collatz conjecture sequence.
  
    """
    ### BEGIN SOLUTION
    # Input validations
    if n <= 0: raise ValueError("n must be positive integer")
    if type(n) != int: raise ValueError("n must be positive integer")
    # Initialize variables
    i=1
    currterm=n
    # Define initial output list
    myseq=list(range(1))
    # Assign initial value of sequence
    myseq[0]=n
    # Iterate until current term equals to 1
    while currterm!=1:
        # Even???
        if (myseq[i-1]%2==0):
            currterm=int(myseq[i-1] / 2)
            myseq.append(currterm)
            # append to list
            i = i+1
        # odd
        else:
            currterm=myseq[i-1]*3 + 1
            myseq.append(currterm)
            # append to list
            i = i+1
    # Then return list
    return(myseq)
            
    
    ### END SOLUTION
```

```python
Collatz(11)
```

```python
def is_prime(num):
    """check if the input number is prime or not.
    A positive integer greater than 1 which has no other factors except 1 and the number itself is called a prime number.
    in your solution, you can consider num is already verified as a positive integer!
    """
    
    if(type(num)!=int):
        raise VauleError('input is not an integer')
    if(num<=0):
        raise VauleError('input is not an integer')
    ### BEGIN SOLUTION   
    # If num is one, return false
    if num == 1:
        return False
    else:
        # iterate over all possible factors of nunm
        for i in range(2,num):
            # If remainder is zero (dividing by iterator), return false
            if num%i==0:
                return False
        # If you finish iterating without returning, then it must be prime
        return True
    ### END SOLUTION
```

```python
is_prime(11)
```

```python
def TwoSum(nums, target):
    """
    :type nums: List[int]
    :type target: int
    :rtype: List[int]
    """
    ### BEGIN SOLUTION
    # Let's loop through the lenght of the list
    ### Note: if we don't use len, range, iterable will not be index
    for i in range(len(nums)):
        # Now let's add each item against every other item
        # Starting from next index position to the end
        for j in range(i+1,len(nums)):
            print(j)
            if (nums[i]+nums[j])==target:
                return [i,j]       
    ### END SOLUTION    
```

```python
def TwoSum2(nums, target):
    """
    :type nums: List[int]
    :type target: int
    :rtype: List[int]
    """
    ### BEGIN SOLUTION
    # Let's loop through the lenght of the list
    ### Note: if we don't use len, range, iterable will not be index
    for i in range(len(nums)):
        # Now let's see use conditional instead of loop
        # Is the difference somewhere in the rest of list?
        if (target-nums[i]) in nums[i+1:]:
            # If so, index is i and....
            # index of the first value of list that meets condition
            # To make index relative to start of nums list add (i+1)
            return ([i,nums[i+1:].index(target-nums[i])+i+1]) 
    ### END SOLUTION    
```

```python
timeit(TwoSum2([3,2,4],6))
```

```python
def myreverse(x):
    ''':type x: int
       :rtype: int
    '''
    ### BEGIN SOLUTION
    # Take absolute value of input and convert to string
    nx=str(abs(x))
    # Reverse order of characters and convert back to integer
    rx=int(nx[::-1])
    # if integer is greater than 2**31-1, return 0
    if rx > (2**31-1):
        return 0
    else:
        # Otherwise, check if original value was positive
        # if yes, return reversed integer
        if x >=0:
            return rx
        # Otherwise, return negated reverse integer
        else:
            return -rx
    ### END SOLUTION
```

```python
def longestCommonPrefix(strs):
    """
    :type strs: List[str]
    :rtype: str
    """
    ### BEGIN SOLUTION
    # Initialize variables
    smallest_word=[]
    smallest_word_len=None
    # Iterate to find smallest word and its lenght
    for word in strs:
        # If either the first or smallest
        if ((smallest_word_len==None) or (len(word)<smallest_word_len)):
            smallest_word=word
            smallest_word_len=len(word)
    
    # Now initialize common prefix variable
    common_prefix=""
    # Iterate across all characters in smallest word
    for i in range(smallest_word_len):
        # Compare each cosecutive character to each word on list
        for word in strs:
            # If it does not start common prefix return
            #print(smallest_word[:(i+1)])
            if not (word.startswith(smallest_word[:(i+1)])):
                return common_prefix
        # Otherwise, redefine common prefix to include extra character
        common_prefix=smallest_word[:(i+1)]
    # return common prefix value
    return common_prefix
    ### END SOLUTION
```

```python
longestCommonPrefix(['what','where','who',"flow","flight"] )
```
