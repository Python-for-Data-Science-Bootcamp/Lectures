## Comparison Operators

```python
1>2           # Comparison Operators < > <= >= != ==
```

```python
x = 6
if x == 5 : 
    print('Equals 5')
if x > 4 : 
    print('Greater than 4')
if  x >= 5 :
    print('Greater than or Equals 5')
if x < 6 : 
    print('Less than 6') 
if x <= 5 :
    print('Less than or Equals 5')
if x != 6 :
    print('Not equal 6')
```

```python
x = 5                               
if x > 2 :
    print('Bigger than 2')
    print('Still bigger')
print('Done with 2')

for i in range(5) :
    print(i)
    if i > 2 : 
        print('Bigger than 2')
    print('Done with i', i)
print('All Done') 
```

## Nested Decisions

```python
x = 1                       #Nested Decisions   #slides page 11
if x > 1 :
    print('More than one')
    if x < 100 : 
        print('Less than 100') 
print('All done')
```

## Two-way Decisions

```python
x = 1                  #Two-way Decisions
if x > 2 :
    print('Bigger')
else :
    print('Smaller')
print('All done')
```

## Multi-way Decisions: In these statements, order matters as questions are checked in order.  

From Py4E slides page 17
![condition](condition4.png)

```python
x=1
if x < 2 :          #Multi-way
    print('small')
elif x < 10 :
    print('Medium')
else :
    print('LARGE')
print('All done')
```

```python
x=-10
if x < 2 :                 #Multi-way Puzzles
    print('Below 2')
elif x >= 2 : 
    print('Two or more')
else :
    print('Something else')
```

## Try and Except: Used to add graceful failure (Python detects error as it tries to execute the instruction) and avoid traceback

```python
astr = '123'
istr = int(astr)
print('Second', istr)
```

```python
astr = 'Luis'
istr = int(astr)
print('Second', istr)
```

### Input Command: Built-in function to instruct a Python program to read input data from the keyboard.

```python
# You can use the raise keyword to signal that the situation is exceptional 
# to the normal flow and you want execution to stop (i.e. traceback). For example:

astr = input('type a positive number! ')
istr = int(astr)
if istr < 0:
    raise ValueError("This is a negative number")
else:
    print("This is the number:",istr)
```

```python
### Implementation with input validation without Traceback
try:
    astr = input('type a number! ')
    istr = int(astr)    #try except
    if (istr>0): 
        print('the number is',istr)
    else:
        print("invalid input")
except:
    print('invalid input')
```

```python
while True:
    try:
        astr = input('type a number! ') 
        istr = int(astr)
        break
    except:
        print("wrong input")
```
