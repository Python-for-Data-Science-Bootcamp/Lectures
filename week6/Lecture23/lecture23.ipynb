## #2 Binomial Distribution

Binomial distribution with parameters n and p is the...

## ...discrete probability distribution which counts the number of successes in a sequence of n independent experiments, 

each asking a yes–no question, and each with its own boolean-valued outcome.  

The binomial distribution is frequently used to model the number of successes k which occur when a coin is tossed n times, assuming that on any one toss, the probability that a head occurs is p. 


## Function Information
### numpy.random.binomial

numpy.random.binomial(n, p, size=None)

Draw samples from a binomial distribution.

Samples are drawn from a binomial distribution with specified parameters, n trials and p probability of success where n an integer >= 0 and p is in the interval [0,1]. (n may be input as a float, but it is truncated to an integer in use)
Parameters:	

n : int or array_like of ints

    Parameter of the distribution, >= 0. Floats are also accepted, but they will be truncated to integers.

p : float or array_like of floats

    Parameter of the distribution, >= 0 and <=1.

size : int or tuple of ints, optional

    Output shape. If the given shape is, e.g., (m, n, k), then m * n * k samples are drawn. If size is None (default), a single value is returned if n and p are both scalars. Otherwise, np.broadcast(n, p).size samples are drawn.

Returns:	

out : ndarray or scalar

Drawn samples from the parameterized binomial distribution, where **each sample is equal to the number of successes over the n trials.**

```python
# Create set of binomial samples
data=np.random.binomial(100,0.1,1000) 
# 1000 samples
# 100 trials
# 0.1 probability of success
# Returns for each sample a number of successes in 100 trials
data[:5]
```

```python
# Plot histogram
plt.hist(data,bins=100); 
```

```python
# Mean => 10.08% of success in 100 trials which matches probability of success p=0.1
np.mean(data)
```

## #3 Geometric Distribution

The geometric distribution is either of two discrete probability distributions:

## ...number X of Bernoulli trials needed to get one success 

supported on the set { 1, 2, 3, ... }
    
## number Y = X − 1 of failures before the first success 

supported on the set { 0, 1, 2, 3, ... }

Which of these one calls "the" geometric distribution is a matter of convention and convenience. 


n = number of trials, p = prob. of success, q = prob. of failure

## Simulation Example

If a coin is tossed 10 times, **what is the probability that the first tail will occur after the third toss, given that it has not occurred in the first two tosses?**

```python
# Let us create our random variable
# We'll use randint to generate random integer values of 0 or 1
# 10 rows to represent number of tosses in each experiement
data=np.random.randint(0,2,[10,100000]) 
data[:,:5]
```

```python
# Create our mask to check for two things:
# 1) No tails (i.e. 1) in first two tosses (sum of values equal to zero)
# 2) It is tails (i.e. 1) on the third toss
mask=(np.sum(data[0:2,:],axis=0)==0) & (data[2,:]==1)
mask[:10]
```

```python
# Now, let's calculate how often mask above was true
prob=np.mean(mask)
print("Probability of seeing tail for first time at 3rd toss is {}%"
      .format(np.round(prob*100,2)))
```

## Now, let's calculate the true probability...

 ## Function Information
## numpy.random.geometric

numpy.random.geometric(p, size=None)

    Draw samples from the geometric distribution.

    Bernoulli trials are experiments with one of two outcomes: success or failure (an example of such an experiment is flipping a coin). The geometric distribution models the number of trials that must be run in order to achieve success. It is therefore supported on the positive integers, k = 1, 2, ....

    The probability mass function of the geometric distribution is

    f(k) = (1 - p)^{k - 1} p

    where p is the probability of success of an individual trial.
    Parameters:	

    p : float or array_like of floats

        The probability of success of an individual trial.

    size : int or tuple of ints, optional

        Output shape. If the given shape is, e.g., (m, n, k), then m * n * k samples are drawn. If size is None (default), a single value is returned if p is a scalar. Otherwise, np.array(p).size samples are drawn.

    Returns:	

    out : ndarray or scalar

   **Drawn samples from the parameterized geometric distribution.**


## Create Samples from Geometric Distribution

```python
# Create our geometrically distributed random variable
data2=np.random.geometric(p=0.5, size=100000) 
# It contains number of trials for each experiment to get to success (i.e. tails)
data2
```

```python
# Check if tails occured in 3rd trial (or toss)
mask=(data2==3)
mask[:10]
```

```python
# Now, let's calculate how often mask above was true
prob=np.mean(mask) 
print("True probability of seeing tail for first time at 3rd toss is {}%"
      .format(np.round(prob*100,2)))
```

## Bottom line, true prob. of 12.4% is very close to estimate of 12.44%

```python
# Plot histogram showing frequency of outcomes (i.e. trials before success)
# Success with fewer trials are more frequent
plt.hist(data2,bins=30); 
```

## #4 Poisson Distribution

Discrete probability distribution that expresses the... 

### ...probability of a given number of events occurring in a fixed interval of time or space 

**if these events occur with a known constant rate (lambda) and independently of the time since the last event.** 

The Poisson distribution can also be used for the number of events in other specified intervals such as distance, area or volume. 

$f(k; \lambda t)=\frac{(\lambda t)^k e^{(-\lambda t)}}{k!}$

t intervals, k successes, lambda rate

### The Poisson distribution is used as an approximation to the binomial distribution when the parameters n (# of trials) and p (probability of success) are large and small, respectively. (see example below)

A train bridge exist across the Ulua river in Honduras. Trains arrive at the bridge according to a Poisson process of rate λ = 2 per day. If a train arrives on day 0, find the probability that there will be no trains for the next four days.




λt occurrences in a time interval of length t

Probability of no trains on a given interval is (k=0): (lambda**0) x (exp**-3)/(0!)

Because it is 4 consecutive days without trains, we will be using above formula with the following parameters:

t = 4 (4 days), k = 0 (no trains), lambda = 2 (trains per day)

## #5 Hypergeometric Distribution

The hypergeometric distribution describes...

### probability of k successes in n draws, **without replacement**, from a finite population of size N that contains exactly K objects with that feature, wherein each draw is either a success or a failure. 



**In contrast, the binomial distribution describes the probability of k successes in n draws with replacement.** 

 ## numpy.random.hypergeometric
 
 numpy.random.hypergeometric(ngood, nbad, nsample, size=None)

    Draw samples from a Hypergeometric distribution.

Samples are drawn from a hypergeometric distribution with specified parameters:

* ngood (ways to make a good selection)
* nbad (ways to make a bad selection)
* nsample = number of items sampled, which is less than or equal to the sum ngood + nbad.
   
   Parameters:	

    ngood : int or array_like of ints

        Number of ways to make a good selection. Must be nonnegative.
    nbad : int or array_like of ints

        Number of ways to make a bad selection. Must be nonnegative.
    nsample : int or array_like of ints

        Number of items sampled. Must be at least 1 and at most ngood + nbad.
    size : int or tuple of ints, optional

        Output shape. If the given shape is, e.g., (m, n, k), then m * n * k samples are drawn. If size is None (default), a single value is returned if ngood, nbad, and nsample are all scalars. Otherwise, np.broadcast(ngood, nbad, nsample).size samples are drawn.

    Returns:	

    out : ndarray or scalar

```python
# Let us define our parameters
# 100 good
# 2 bad
# 10 samples
ngood, nbad, nsamp = 100, 2, 10
# ngood = 100 green 
# nbad = 2 blue 
# nsamp = 10 Number of balls sampled
# number of good, number of bad, and number of samples
s = np.random.hypergeometric(ngood, nbad, nsamp, 1000)
s[:10]
# Number of successes (i.e. pulled good) in 10 draws..
```

```python
# Histogram
plt.hist(s);
```

## Excercise
Suppose you have an urn with 15 white and 15 black marbles. If you pull 15 marbles at random, how likely is it that
**12 or more** of them are one color?

```python
# 15 sample items
# 15 good (say black)
# 15 bad (say white)
s = np.random.hypergeometric(15, 15, 15, 100000) # create random variable
# Number of "Good" outcomes in 15 samples
s[:10]
```

```python
# Probability that majority is black? check if s>=12 (successes are more or eq to 12)
# Probability that majority is white? check if s<=3 (successes are less or eq to 3)
# First factor accounts for majority Black
# Second factor accounts for majority White
# Because we want probability that either case occurs, we must add them
prob=(np.mean(s>=12) + np.mean(s<=3))
print("Probability that 12 or more of them are one color is {}%"
      .format(np.round(prob*100,2)))
```

## #6 Continuous Uniform Density

The continuous uniform distribution is the ...

## ...probability distribution of random number selection from the continuous interval between a and b. 

The simplest density function corresponds to the random variable U whose value represents the outcome of the experiment consisting of choosing a real number at random from the interval [a, b].


```python
# Goal: Create uniformly distributed set between 1 and 3
# 100k random samples between 0 and 1 using rand function
data=np.random.rand(100000) 
data[:10]
```

```python
# To make it uniformly distributed between 1 and 3 
data=data*2+1 
# Multiply by two and add 1
plt.hist(data,bins=100);
```

```python
np.mean(data) # Calculate mean
```

## In closing, we reviewed the following distributions:

- Discrete Uniform - discrete outcomes equally likely (e.g dice)
- Binomial - successes in n trials with replacement (e.g. how many heads in 10 tosses)
- Geometric - bernoulli trials needed to get to success (e.g how many tosses to get to a head)
- Poisson - events happening within an interval of time (e.g. likelihood of 3 trains arriving on third day, bombs hitting a square)
- Hypergeometric - k successes in n trials without replacement (e.g. pulling 3 black marbles from a jar)
- Continous Uniform - infinite outcomes within interval (e.g. probability of dots falling inside unitary circle)

For each of these, we discussed how to assign probabilities using formulas and simulations
